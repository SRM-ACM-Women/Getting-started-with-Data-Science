#### selecting specific rows
![rows](https://pandas.pydata.org/docs/_images/03_subset_rows.svg)

```python
above_35 = titanic[titanic["Age"] > 35]
above_35.head()
```

```python
# if i want to see Titanic passengers from cabin class 2 and 3.
class_23 = titanic[titanic["Pclass"].isin([2, 3])]
class_23.head(5)
```

#### selecting specific rows and columns from the dataframe
![rowAndColumn](https://pandas.pydata.org/docs/_images/03_subset_columns_rows.svg)

```python
# selecting the names of the passengers older than 35 years.
adult_names = titanic.loc[titanic["Age"] > 35, "Name"]
adult_names.head()
```

```python
# selecting rows 10 till 25 and columns 3 to 5.
titanic.iloc[9:25, 2:5]
```
